# Project-5 data pre-processing


We will be processing and enriching data which will be used on Project-5.ipynb for moral relevancy classification task using BERT-based framework. 

```python
import json
import numpy
import pandas
```

```python
# This is the same data that we used for moral rating prediction task using SBERT and FastText.

with open("../data/structured_data_full.json", "r") as f:
    data = json.load(f)
```

```python
def count_sentences(data):
    total_sentences = 0
    moral_sentences = 0
    non_moral_sentences = 0
    for movie_name, characters in data.items():
        for char_name, sent_types in characters.items():
            for type, sentences in sent_types.items():
                if type == "moral":
                    total_sentences += len(sentences)
                    moral_sentences += len(sentences)
                if type == "non_moral":
                    total_sentences += len(sentences)
                    non_moral_sentences += len(sentences)

    return total_sentences, moral_sentences, non_moral_sentences

sentence_count = count_sentences(data)
print(f"Total number of sentences: {sentence_count[0]}")
print(f"Total moral sentences: {sentence_count[1]}")
print(f"Total non-moral sentences: {sentence_count[2]}")
```

```python
def flatten_data_to_dataset(data):
    flattened = []
    for movie_name, characters in data.items():
        for char_name, sent_types in characters.items():
            for sent_type, sentences in sent_types.items():
                if sent_type == "moral":
                    for sentence in sentences:
                        flattened.append({"text": sentence, "label": 1})
                elif sent_type == "non_moral":
                    for sentence in sentences:
                        flattened.append({"text": sentence, "label": 0})
    return flattened

flattened_data = flatten_data_to_dataset(data)
```

```python
with open("../data/data_for_project_5.json", "w") as f:
    json.dump(flattened_data, f, indent=4)
```
