# Data Exploratory

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open('..//data//structured_data.json', 'r') as file:
    structured_data = json.load(file)
```

```python
num_characters = 0
num_movies = len(structured_data)

for movie, characters in structured_data.items():
    for character, sentences in characters["characters"].items():
        num_characters += 1

print(f'Total number of movies: {num_movies}')
print(f'Total number of characters: {num_characters}')
```

```python
# Let's load the rating data

with open("../data/new_dialogue.json", "r") as file:
    dialogue_data = json.load(file)
```

```python
# We'd like to count the number of sentences from characters with more than 50 and 100 sentences

num_sentences_100 = 0
num_sentences_50 = 0
for movie, data in dialogue_data.items():
    for character, sentences in data.items():
        if len(sentences) > 100:
            num_sentences_100 += len(sentences)
        if len(sentences) > 50:
            num_sentences_50 += len(sentences)

print(f'Total number of sentences from characters with more than 100 sentences: {num_sentences_100}')
print(f'Total number of sentences from characters with more than 50 sentences: {num_sentences_50}')
```

```python
with open("..//data//new_moral_data_full.json") as f:
    moral_dialogue_data = json.load(f)["moral_dialogue"]

num_sentences_100 = 0
num_sentences_50 = 0

for movie, data in moral_dialogue_data.items():
    for character, sentences in data.items():
        if len(sentences) > 100:
            num_sentences_100 += len(sentences)
        if len(sentences) > 50:
            num_sentences_50 += len(sentences)

print(f'Total number of sentences from characters with more than 100 sentences: {num_sentences_100}')
print(f'Total number of sentences from characters with more than 50 sentences: {num_sentences_50}')
```

```python
import json 

with open("../data/structured_data_full.json", "r") as file:
    structured_data_full = json.load(file)
```

```python
total_sentences_by_type_all_characters = {}
char_num = 0

for movie, characters in structured_data_full.items():
    for character, data in characters.items():
        char_num += 1
        for sentence_type, sentences in data.items():
            total_sentences_by_type_all_characters[sentence_type] = total_sentences_by_type_all_characters.get(sentence_type, 0) + len(sentences)

print(total_sentences_by_type_all_characters)
```
